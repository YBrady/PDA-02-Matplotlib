# 2. Two Plots on One Set of Axes #

```python
import numpy as np
x = np.arange(0.0, 10.0, 0.01) # Makes a range (can handle decimals)
y = 3.0 * x + 1.0 # Just making y a function of x

noise = np.random.normal(0.0, 1.0, len(x)) # Generates noise

plt.plot(x,y + noise, "r.") # Red dot
plt.plot(x,y,"b-") # Blue line
plt.show() # Cleans the slate with respect to adding plots to the plot
```

## 3. Titles, Legends and Labels ##

```python
x = np.arange(0.0, 10.0, 0.01) # Makes a range (can handle decimals)
y = 3.0 * x + 1.0 # Just making y a function of x

noise = np.random.normal(0.0, 1.0, len(x)) # Generates noise

plt.plot(x,y + noise, "r.", label = "Actual") # Red dot
plt.plot(x,y,"b-", label = "Model") # Blue line

plt.title("Simple Plot")
plt.xlabel("Weight")
plt.ylabel("Mass")
plt.legend()

plt.show() # Cleans the slate with respect to adding plots to the plot
```

## 4. Histograms (1-D Data) ##

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(0.0, 1.0, 1000)

plt.hist(x)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(0.0, 1.0, 1000)

plt.hist(x, bins = 20) # Make it more granular

plt.show()
```

## 5. Side-by-Side Plots ##

```python
import matplotlib.pyplot as plt
import numpy as np

plt.subplot(1,2,1) # No of rows, No of columnns, This is the first one in it
x = np.random.normal(0.0, 1.0, 1000) # Normal distribution (bell curve)
plt.hist(x) 

plt.subplot(1,2,2) # No of rows, No of columnns, This is the second one in it
x = np.random.uniform(-3.0, 3.0, 1000) # Uniform distribution
plt.hist(x) 

plt.show()
```

## 6. Other Plots ##

```python
import matplotlib.pyplot as plt
import numpy as np

# Plotting four parameters in a 2-D plot
x = np.random.uniform(0.0, 10.0, 100) # x-values
y = np.random.uniform(0.0, 100.0, 100) # y-values
z = np.random.normal(100.0, 40.0, 100) # sizes
c = np.random.randint(0, 20, 100) # colours

plt.scatter(x, y, c = c, s = z) # x, y, colour, size

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-2.0 * np.pi, 2.0 * np.pi, 0.1)

plt.plot(x, np.sin(x), "g.")
plt.plot(x, np.cos(x), "b.")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1.0, 10.0, 0.1)

plt.plot(x, x**2, "g-", label = "x^2")
plt.plot(x, x**3, "b-", label = "x^3")
plt.plot(x, x**4, "r-", label = "x^4")
plt.plot(x, 2**x, "k-", label = "2^x")

plt.legend()

plt.show()
```
